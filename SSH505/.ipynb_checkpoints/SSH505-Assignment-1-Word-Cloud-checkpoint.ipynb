# Import packages

```python
# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
from PIL import Image # converting images into arrays
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
question1 = open('question1.txt', 'r', encoding="utf8").read()
question2 = open('question2.txt', 'r', encoding="utf8").read()
question3 = open('question3.txt', 'r', encoding="utf8").read()
question4 = open('question4.txt', 'r', encoding="utf8").read()
question5 = open('question5.txt', 'r', encoding="utf8").read()

wordcloud_mask1 = np.array(Image.open('people.png'))
wordcloud_mask2 = np.array(Image.open('values.png'))
wordcloud_mask3 = np.array(Image.open('read.png'))
wordcloud_mask4 = np.array(Image.open('hope.png'))
wordcloud_mask5 = np.array(Image.open('future.png'))


questions = [question1, question2, question3, question4, question5]
masks = [wordcloud_mask1, wordcloud_mask2, wordcloud_mask3, wordcloud_mask4, wordcloud_mask5]
```

```python
stopwords = set(STOPWORDS)
```

```python
stopwords.update(['people', 'someone', 'Someone', 'person','say','consider','thing','come','things',
                  'value','values','may','might','important','really','much','know','read','reader','reading',
                  'hope','future'])

for i in range(len(questions)): 
    wordcloud = WordCloud(background_color = 'white',
                    max_words = 200,
                    mask=masks[i],
                    width=1000, 
                    height=500, 
                    stopwords=stopwords,
                    contour_width=10,
                      contour_color = 'black'
                     ).generate(questions[i])
    wordcloud.to_file("question{}.png".format(i))

    plt.figure( figsize=(30,15) )
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```
